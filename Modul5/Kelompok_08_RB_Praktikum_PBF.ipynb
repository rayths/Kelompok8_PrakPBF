## Normalize Data Decorator

### Import Library

```python
import pandas as pd
from functools import wraps
```

### Fungsi Decorator Normalisasi Data

```python
def normalize_data(func):

  @wraps(func)
  def decorated_function(df, *args, **kwargs):

    data = df.select_dtypes(include='number')
    normalized_data = (data - data.min()) / (data.max() - data.min())
    df.update(normalized_data)
    return func(df)

  return decorated_function
```

### Import Dataset

https://drive.google.com/file/d/10uoQgxOvjM_fs6ow2e1DkAKPqbDFejWg/view?usp=drive_link

```python
df = pd.read_csv("insurance.csv")
```

### Dataset

```python
df.sample(5)
```

### Penggunaan Decorator Normalisasi Data

#### Dataset Setelah dinormalisasi

```python
@normalize_data
def dataset(data):
  return data.sample(5)

dataset(df)
```

#### Statistika Deskriptif Data

```python
@normalize_data
def statistika_deskriptif(data):
  return data.describe()

statistika_deskriptif(df)
```
